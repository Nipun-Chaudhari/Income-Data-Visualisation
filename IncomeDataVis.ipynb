# DATASET CHANGED

"Income2.csv" is same as "Income1.csv" but an extra column "Seniority" is added.

```python
df2 = pd.read_csv("Income2.csv")
print(df2.head(),'\n',df2.tail())

cols = ['Unnamed: 0']
df2 = df2.drop(cols, axis = 1)

print(df2.describe())
sns.scatterplot(data = df2, x = "Education", y = "Income").set(title = "Scatterplot")
sns.lmplot(data = df2, x = "Education", y = "Income").set(title = "Regression Plot")
```

Data in "Income2.csv" is more distributed, this can be infered from "Education-Income" scatterplot. Therefore regression line fit over education and income is not sharp. But it still shows that there is a direct relation between 
education and income.

```python
sns.displot(df2, x = "Seniority")
```

```python
sns.scatterplot(data = df2, x = "Seniority", y = "Education", color = "green")
```

```python
sns.scatterplot(data = df2, x = "Seniority", y = "Income", color = "blue")
```

```python
sns.lineplot(data = df2, x = "Seniority", y = "Education", color = "red")
```

```python
sns.lineplot(data = df2, x = "Income", y = "Seniority", color = "red")
```

```python
sns.lmplot(x = "Education", y = "Seniority", x_estimator = np.mean, data = df2).set(title = "Education-Seniority Plot")
sns.lmplot(x = "Income", y = "Seniority", x_estimator = np.mean, data = df2).set(title = "Income-Seniority Plot")
```

Data values of "Seniority" are more scattered. When plot against "Education" and "Income" it does not show specific trend. Above plots show that education does not affect seniority, but most seniors have more income.
